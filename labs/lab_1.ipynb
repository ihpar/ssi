```python
def plot_spectrogram(signal, title):
    stft = librosa.stft(signal)
    spec = librosa.amplitude_to_db(np.abs(stft))
    plt.figure(figsize=(12, 8))
    librosa.display.specshow(spec, y_axis="log")
    plt.colorbar(format="%+2.0f dB")
    plt.title(title)
    plt.xlabel("t")
    plt.show()
```

```python
plot_spectrogram(piano, "Piyano A 3 (220 Hx)")
```

```python
plot_spectrogram(zil, "Zil")
```

```python
plot_spectrogram(davul, "Davul")
```
